**Import pandas package and loading two datasets**

````python
import pandas as pd
df2019 = pd.read_csv('Compustat_fy2019.csv', parse_dates = ['datadate'])
df2018 = pd.read_csv('Compustat_fy2018.csv', parse_dates = ['datadate'])
````

---
## Appending (Concatenating) Data
```Python
df2018_2019 = pd.concat([df2019,df2018], ignore_index = True)
```
- Appending the 2018 dataset to 2019 the dataset.
- How the concatenating works: the head of df2018 (second df) is attached to the tail of df2019 (first df); Python will automatically attach the columns from both datasets based on the variable name. If there are variables (SIC code in this case) in df2018 but not in df2019, these variables will also be saved in additional columns in the final dataset (df2018_2019).
- Setting the parameter (ignore_index) true will reorganize/reset the index for the final dataset. Python with essentially re-count your data beginning at 0 and ending at the end of the second dataframe.
- You can visualize this as stacking two dataframes on top of each other

````python
df2018_2019 = pd.concat([df2019,df2018], ignore_index = True)
````

## Merging Data
- In some situations, the variables you need for your analyses could come from multiple independent datasets. In these scenarios, you would prefer to merge these datasets together to get the variables you need in the same place. 
- Merging datasets require bridges to connect the observations between two datasets. We usually call these bridges identifiers. As the name suggests, identifiers help identify the observations or the unit of analysis that you are interested in. Some examples of identifiers could include people's names, company names or tickers, and locations.

## A Small Example

````python
dfage = pd.DataFrame({'name': ['Jack', 'Mary'],
                    'age': [35, 18]},
                   columns=['name', 'age'])
dfhome = pd.DataFrame({'name': ['Mary', 'Li'],
                    'hometown': ['New Orleans, LA', 'Miami, FL']},
                   columns=['name', 'hometown'])
````
