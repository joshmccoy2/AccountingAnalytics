```python
import pandas as pd
```

```python
# Load into a DataFrame
worker_df = pd.read_csv('Worker.csv')
payitem_df = pd.read_csv('PayItem.csv')
job_df = pd.read_csv('Job.csv')
worker_itemized_amount_df = pd.read_csv('Worker_Itemized_Amount.csv')
```

```python
worker_df.head()
```

```python
df_merged_1 = pd.merge(worker_itemized_amount_df, worker_df, on='WorkerID', how='left')
df_merged_1
```

```python
df_merged_2 = pd.merge(df_merged_1, payitem_df, on='PayItemID', how='left')
df_merged_2
```

```python
df_final_merged = pd.merge(df_merged_2, job_df, on='WorkCode', how='left')
df_final_merged
```

```python
filename = 'merged_dataframe.csv'
df_final_merged.to_csv(filename, index=False)
```

```python
#Training Expenses
training_items = ['training - basePay', 'training - airFare', 'training - meals', 'training - lodging']
training_expenses = df_final_merged[df_final_merged['PayItemName'].isin(training_items) & (df_final_merged['Status'] == 'IC')]
total_training_expenses = training_expenses['ItemizedAmount'].sum()

#Travel Expenses
travel_items = ['mileage', 'parking', 'training - airFare']
travel_expenses = df_final_merged[df_final_merged['PayItemName'].isin(travel_items) & (df_final_merged['Status'] == 'IC')]
total_travel_expenses = travel_expenses['ItemizedAmount'].sum()

#Office Expenses
office_items = ['officeEquip - laptop', 'officeEquip - scanner', 'officeEquip - tablet']
office_expenses = df_final_merged[df_final_merged['PayItemName'].isin(office_items) & (df_final_merged['Status'] == 'IC')]
total_office_expenses = office_expenses['ItemizedAmount'].sum()

# Print results
total_training_expenses
```

```python
from datetime import datetime
```

```python
# Converting to datetime
df_final_merged['HireDate'] = pd.to_datetime(df_final_merged['HireDate'])
df_final_merged['EndDate'] = pd.to_datetime(df_final_merged['EndDate'])

# if NA make current date
current_date = pd.to_datetime('2016-08-01')
df_final_merged['EndDate'].fillna(current_date, inplace=True)


ic_df = df_final_merged[df_final_merged['Status'] == 'IC']

#length of employment
ic_df['LengthOfEmployment'] = (ic_df['EndDate'] - ic_df['HireDate']).dt.days

# average length of employment for ICs
average_length_of_employment_ic = ic_df['LengthOfEmployment'].mean()

average_length_of_employment_ic
```

This also makes us further confident that the companty is not properly classifying the employees because the average length of employment is 318 days which is a long term worker
